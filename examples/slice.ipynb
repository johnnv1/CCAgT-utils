# Slicing images


## Using the entrypoint

----- 

Use on the terminal:

```shell
CCAgT-utils create-subdataset <SUBDATASET_NAME> --original <PATH_FOR_ORIGINAL_DATASET> --output <PATH_TO_WRITE_THE_SUBDATASETS> --slice-images 4 2
```

using the argument `--slice-images` will slice the images in 4 parts horizontally and 2 parts vertically

## Using the function directly 

````python
from CCAgT_utils import slice
import tempfile
````

````python
image_path = '../data/samples/images/C_13509_-270720_116640.jpg'
output_dir = tempfile.TemporaryDirectory()
````

````python
# Slicing the image into 4 parts horizontally and 2 parts vertically
qtd_slice = slice.image(image_path, output_dir.name, h_quantity=4, v_quantity=2)
````
